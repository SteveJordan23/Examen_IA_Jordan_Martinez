### Ejercicio cuatro: Segun el ejercicio tres, que tanto aumento o disminuyo el precio de cada item-variety 

Ejemplo de como se debe de mostrar su respuesta

```python
pd.DataFrame({
    "item" : ['Manzana','Manzana','Manzana']
    , "variety" : ['Gala','Gala','Gala']
    , "year" : [2020,2021,2022]
    , "last_price" :  [2.7,2.5,3.2]
    , "delta" : [0, (2.5-2.7), (3.2-2.5)]
})
```

Segun nuestro ejemplo dummy para 2021 estuvo 0.2 centavos mas barato y en 2022 estuvo 0.7 centavos caro

```python
vdf['date']= pd.to_datetime(vdf['date'])
vdf['Year'] = vdf['date'].dt.year 
vdf['Month'] = vdf['date'].dt.month
```

```python
last_price = vdf.loc[ : , ['item', 'variety', 'Year', 'price']].drop_duplicates()
last_price = last_price.sort_values(['item', 'variety', 'Year'], ascending = True)
```

```python
last_price = vdf.loc[ :, ["item", "variety", 'Year', 'price'] ].groupby(by = [ "item", "variety", 'Year'] ).shift(1)
```

```python
last_price['New_price'] = vdf.groupby(["item", "variety", 'Year'], as_index=False)['price'].shift(-1)
last_price = last_price.loc[~pd.isna(last_price['price']) , :]
```

```python
last_price.insert(0, 'delta', (last_price['price'] - last_price['New_price']))
```

### Ejercicio seis: muestre por category,item,variety que tanto % ha aumentado su precio

Ejemplo: si para las manzanas gala el primer precio que se registro fue 2.00 y el ultimo precio que se registro es 4.00 este aumento un 100%

```python
pd.DataFrame({
    "category": ["fruta","fruta"]
    ,"item" : ['Manzana','Manzana']
    , "variety" : ['Gala','egremont_russet']
    , "percentage-change-between-first-and-last-price" : [200,150]       
})
```

```python
last_price['first_price'] = vdf.loc[ :, ["category", "item", "variety", 'price'] ].groupby(by = ["category", "item", "variety"], as_index=False ).first()
```

```python
last_price['last_price'] = vdf.loc[ :, ["category", "item", "variety", 'price'] ].groupby(by = ["category", "item", "variety"], as_index=False ).last()
```

```python
last_price.insert(0, 'Porcentaje de cambio', (last_price['first_price'] / last_price['last_price']).sum()*100)
```
